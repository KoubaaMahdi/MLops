```python
# Updated function to visualize the relationship with a specific column
def visualize_relationship(dataframe, target_column, feature_column):
    """
    Function to visualize the relationship between the target column and a specific feature column.

    Parameters:
    - dataframe (pd.DataFrame): The dataset.
    - target_column (str): The target column to study relationships.
    - feature_column (str): The specific column to visualize with the target column.

    Returns:
    - None
    """
    plt.figure(figsize=(10, 6))
    
    if dataframe[feature_column].dtype == "object" or dataframe[feature_column].nunique() < 10:
        # Categorical feature: use a countplot with hue
        sns.countplot(data=dataframe, x=feature_column, hue=target_column, palette="viridis")
        plt.title(f"Relationship between {target_column} and {feature_column}")
    else:
        # Numerical feature: use a boxplot
        sns.boxplot(data=dataframe, x=target_column, y=feature_column, palette="viridis")
        plt.title(f"Distribution of {feature_column} by {target_column}")
    
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Example usage
visualize_relationship(data, "History of Mental Illness", "Age")
#visualize_relationship(df, "History of Mental Illness", "Marital Status")
```

The boxplot illustrates the distribution of age for individuals grouped by their history of mental illness ("Yes" or "No"). Key observations are as follows:

1. *Median Age*: The median age is relatively similar for both groups, indicating that the central tendency of age does not significantly differ between individuals with and without a history of mental illness.

2. *Interquartile Range (IQR)*: Both groups have a similar IQR, meaning that the middle 50% of ages are distributed over a comparable range for both groups.

3. *Outliers*: There are no significant outliers in either group, suggesting that the age distributions are consistent within their respective ranges.

4. *Overall Spread*: The overall age spread (minimum to maximum) is nearly identical for both groups, implying that age does not show a strong differentiating pattern based on the history of mental illness.

*Conclusion*: Age does not appear to have a distinct or significant relationship with whether an individual has a history of mental illness, as the distributions across the two groups are quite similar.
