```python
df = pd.read_csv("../data/depression_data.csv")
df.head()
```

```python
import deepchecks
deepchecks.__version__
```

```python
from deepchecks.tabular.suites import full_suite
suites=full_suite()
suites
```

```python
cat_features= df.select_dtypes(include=['object']).columns.tolist()
cat_features.remove('History of Mental Illness')
cat_features.remove('Name')
cat_features
```

```python
from deepchecks.tabular import Dataset
dataframeTest=Dataset(df=df,label='History of Mental Illness',cat_features=cat_features)
dataframeTest
```

```python
suites.run(dataframeTest)
```

```python
    from deepchecks.suites import data_integrity
```

```python
integrity_test=data_integrity()
integrity_test
```

```python
result_integrity_test=integrity_test.run(dataframeTest)
result_integrity_test
```

```python
htmlSave=result_integrity_test.save_as_html('result_integrity_test')
```

```python
dfcopy=df.copy()
dfcopy.drop(['Name'] , axis=1, inplace=True)
#get categorical features
cat_features= dfcopy.select_dtypes(include=['object']).columns.tolist()
cat_features.remove('History of Mental Illness')
dataframeTest=Dataset(df=df,label='History of Mental Illness',cat_features=cat_features)
dataframeTest
```

```python
from sklearn.model_selection import train_test_split
df_train,df_test=train_test_split(dfcopy,test_size=0.2)
```

```python
print(f"df_train:{df_train.shape}")
print(f"df_test:{df_test.shape}")
```

```python
#transformation
ds_train=Dataset(df=df_train,label='History of Mental Illness',cat_features=cat_features)
ds_test=Dataset(df=df_test,label='History of Mental Illness',cat_features=cat_features)
data_distribution_test=suites.run(train_dataset=ds_train,test_dataset=ds_test)
data_distribution_test
```
